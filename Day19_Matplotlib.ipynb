```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
# Creating Pie Charts

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as ny

y = np.array([35, 25, 25, 15])
mylabels = ["apple", "banana", "cherries", "dates"]

plt.pie(y, labels= mylabels)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels= mylabels, startangle= 90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apple", "Cherries", "Dates", "Bananas"]
myexpode = [0.2, 0, 0, 0]
mycolor = ["Red", "purple", "Green", "Yellow"]

plt.pie(y, labels= mylabels, startangle= 90, explode= myexpode, shadow= True, colors= mycolor)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apple", "Banana", "Dates", "Pears"]

plt.pie(y, labels= mylabels, startangle= 90)
plt.legend(title = 'Fruit', loc= 1)
plt.show()
```
